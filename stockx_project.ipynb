```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use('ggplot')
import seaborn as sns
```

```python
stockx_df=pd.read_csv('~/Desktop/stockx/stockx.csv')
```

```python
#clean data
stockx_df=stockx_df[~stockx_df['highest_bid'].isnull()] 
stockx_df=stockx_df[~stockx_df['avg_sale_price'].isnull()]
```

```python
#add total sales column
stockx_df["total_sales"]=stockx_df["avg_sale_price"]*stockx_df["number_of_sales"]
```

```python
stockx_df.head()
```

```python
# add trend column 
stockx_df['trend']=np.where(stockx_df['highest_bid']>stockx_df['avg_sale_price'],'up','down')
```

```python
stockx_df
```

```python
#extract model number from style column
stockx_df['model']=stockx_df['style'].str.split('-').str[0]
```

```python
stockx_df
```

```python
#extract release year from release date
stockx_df['release_year']=stockx_df['release_date'].str.split('/').str[2]
stockx_df
```

```python
#extract base color form color
stockx_df['base_color']=stockx_df['color'].str.split('/').str[0]
stockx_df
```

```python
#extract model name from product title

stockx_df['model_category']=stockx_df['product_title'].apply(lambda x: ' '.join(str(x).split(' ')[0:4]))
```

```python
stockx_df
```

```python
#create a dataframe of information of model_category

model_df=stockx_df.groupby(['model_category']).agg(unique_sub_category=("product_title","nunique"),
                                        avg_retail_price=("retail_price",'mean'),
                                        std_retail_price=("retail_price",'std'),
                                        avg_sale_price=("avg_sale_price",'mean'),
                                        max_sale_price=("avg_sale_price",'max'),
                                        min_sale_price=("avg_sale_price",'min'),
                                        std_sale_price=("avg_sale_price",'std'),
                                        sum_total_sales=("total_sales",'sum'),
                                        avg_price_premium=("price_premium",'mean'),
                                        std_price_premium=("price_premium",'std'),
                                        avg_volatility=('volatility','mean'),
                                        std_avg_volatility=('volatility','std'))
                    
```

```python
#sort value by std price premium

model_df.sort_values(by="std_price_premium", ascending=False).head(15)
```

```python
#round value of model_df

model_df=round(model_df)
model_df
```

```python
#remove data that has null value in the std_sale_price column 
#create minimum price premium column
model_df=model_df[~model_df['std_sale_price'].isnull()]
model_df['minimum_price_premium']=((model_df['min_sale_price']-model_df['avg_retail_price'])/model_df['avg_retail_price'])*100

#create another dataframe sorted by minimum price premium (only minimum price premium > 100% )
model_df_list=model_df[model_df['minimum_price_premium']>100].sort_values(by="minimum_price_premium", ascending=False).head(10)
```

```python
model_df_list
```

```python
#create a list of top 10 minimum price premium
top_10_premium=model_df_list.index.tolist()
top_10_premium
```

```python
#plot model category with highest price premium
model_df.groupby('model_category')['minimum_price_premium'].mean().sort_values(ascending=False).head(10).plot.barh(color='k').invert_yaxis()
plt.xlabel('Minimum Price Premium (%)')
plt.ylabel('Model Category')
```

```python
#boxplot 

stockx_df.loc[stockx_df['model_category'].isin(top_10_premium)].boxplot(by='model_category',column=['price_premium'],figsize=(12,8))
plt.xticks(rotation=45, ha="right")
plt.title(' ')
plt.xlabel('Model Category')
plt.ylabel('Price Premium(%)')
stockx_df.loc[stockx_df['model_category'].isin(top_10_premium)].boxplot(by='model_category',column=['avg_sale_price'],figsize=(12,8))
plt.xticks(rotation=45, ha="right")
plt.title(' ')
plt.xlabel('Model Category')
plt.ylabel('Average Sale Price')
stockx_df.loc[stockx_df['model_category'].isin(top_10_premium)].boxplot(by='model_category',column=['retail_price'],figsize=(12,8))
plt.xticks(rotation=45, ha="right")
plt.title(' ')
plt.xlabel('Model Category')
plt.ylabel('Retail Price')

stockx_df.loc[stockx_df['model_category'].isin(top_10_premium)].boxplot(by='model_category',column=['number_of_sales'],figsize=(12,8))
plt.xticks(rotation=45, ha="right")
plt.title(' ')
plt.xlabel('Model Category')
plt.ylabel('Number of Sales')


```
